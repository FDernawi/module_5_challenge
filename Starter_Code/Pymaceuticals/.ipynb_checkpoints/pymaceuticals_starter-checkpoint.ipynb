### Analysis

The analysis indicates that Capomulin and Ramicane are promising treatments for reducing tumor volumes, with Capomulin showing consistent results in decreasing tumor size over time. There is a positive relationship between mouse weight and tumor volume, and some variability in treatment response exists across the different regimens, as highlighted by the presence of outliers. This information can help guide further research and optimization of drug dosage to improve efficacy and consistency in treatment outcomes. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="inner")

# Display the data table for preview
merged_data
```

```python
# Checking the number of mice.
#create a variable for checking unique values in column
#new variable = pandas library . function ( referenced dataframe [referenced cell header] )
unique_mouse_id = pd.unique(merged_data['Mouse ID'])
# initiate the variable at 0 to start the count
# https://www.geeksforgeeks.org/python-get-unique-values-list/?ref=header_outind
mouse_unique = 0 
for item in unique_mouse_id:
    mouse_unique +=1
#printing the number of unique values within the group
print(f'Number of unique values present in the Mouse ID is: {mouse_unique}')
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
unique_timepoint = pd.unique(merged_data['Timepoint'])
timepoint_unique = 0
for item in unique_timepoint:
    timepoint_unique +=1
print(f'Number of unique values present in the Timepoint is: {timepoint_unique}')

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
    #https://www.geeksforgeeks.org/find-duplicate-rows-in-a-dataframe-based-on-all-or-selected-columns/?ref=header_outind
duplicate_mice = merged_data[merged_data.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
#use the loc function to select the first column

duplicate_mice = merged_data.loc[merged_data["Mouse ID"] == "g989", :]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#https://www.geeksforgeeks.org/python-pandas-dataframe-drop_duplicates/?ref=header_outind

clean_data = merged_data[merged_data["Mouse ID"] != "g989"]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame. (Reference to line 2 of the code)
#change the values associated with the previous block of code, adjusted to referenced dataframe and new module
#create a variable for checking unique values in column
mice_count = pd.unique(clean_data['Mouse ID'])
# initiate the variable at 0 to start the count
# https://www.geeksforgeeks.org/python-get-unique-values-list/?ref=header_outind
mouse_unique_clean = 0 
for item in mice_count:
    mouse_unique_clean +=1
#printing the number of unique values within the group
print(f'Number of unique values present in the Mouse ID in the mice count dataframe is: {mouse_unique_clean}')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.

    # https://www.geeksforgeeks.org/how-to-calculate-summary-statistics-in-pandas/?ref=header_outind <----calculation summary stats
        # https://www.geeksforgeeks.org/python-pandas-dataframe-aggregate/?ref=header_outind <----- aggregation method 
       # We would group the data by drug regimine so we can calculate the data for tumor volume, create new variable 

regimen_stats = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean ='mean',
    Median ='median',
    Variance = 'var',
    Standard_Deviation = 'std',
    SEM = 'sem'
)    

# Assemble the resulting series into a single summary DataFrame.
    #to do this we would create a new variable that utilizes the pandas library to create a new dataframe that calls on the dataframe created above
#note to self "simpler way to do this would be to simply rename the dataframe above"

summary_stats = pd.DataFrame(regimen_stats)
summary_stats

#note to self "do not print both dataframes"
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoints = clean_data.groupby(["Mouse ID"]) ["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_merge = pd.merge(last_timepoints, clean_data, on=("Mouse ID", "Timepoint"))
last_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# The list of treatments include Capomulin, Ramicane, Infubinol, and Ceftamin
    #simply create a list variable
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Creation of empty lists for the volume data of the treatments
capomulin_vol_data = []
ramicane_vol_data= []
infubinol_vol_data= []
ceftamin_vol_data= []

#creation of for statement to populate the lists above 
# the most efficient method to complete this task would be to group the data accordingly and use the grouped data for next steps
grouped = last_merge.groupby('Drug Regimen')

capomulin_vol_data = grouped.get_group('Capomulin')['Tumor Volume (mm3)'].tolist()
ramicane_vol_data = grouped.get_group('Ramicane')['Tumor Volume (mm3)'].tolist()
infubinol_vol_data = grouped.get_group('Infubinol')['Tumor Volume (mm3)'].tolist()
ceftamin_vol_data = grouped.get_group('Ceftamin')['Tumor Volume (mm3)'].tolist()

# Printing the filled lists to verify
print("Capomulin Volume Data:", capomulin_vol_data)
print("Ramicane Volume Data:", ramicane_vol_data)
print("Infubinol Volume Data:", infubinol_vol_data)
print("Ceftamin Volume Data:", ceftamin_vol_data)


# Before calculating the IQR we need to create a new dataframe with the above parameters

# Calculate the IQR and quantitatively determine if there are any potential outliers.
    # https://www.geeksforgeeks.org/how-to-use-pandas-filter-with-iqr/?ref=header_outind
for drug in treatments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = clean_data.loc[clean_data["Drug Regimen"] == drug, 'Tumor Volume (mm3)']

    # add subset
    tumor_vol_list.append(final_tumor_vol)

    # Calculating the bounds
    q1 = final_tumor_vol.quantile(0.25)
    q3 = final_tumor_vol.quantile(0.75)
    iqr = st.iqr(final_tumor_vol)

    # Determine outliers using upper and lower bounds
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = final_tumor_vol[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]

    # Iterate over each drug in treatments
    # Printing the results for the treatment group
    print(f"Treatment Group: {drug}")
    print(f"  IQR: {iqr}")
    print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"  Outliers: {outliers.tolist()}\n")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.

plt.figure(figsize=(10, 6))
plt.boxplot(tumor_vol_list, labels=treatments)

# Set plot labels and title
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Distribution of Tumor Volume for Each Treatment Group')

# Show the plot
plt.show()

# Reference
#https://www.geeksforgeeks.org/box-plot-in-python-using-matplotlib/?ref=header_outind
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
#there is only one mouse that is treated with capomulin this can cause variance within the data and should be written in the remarks at the end

# Filtering the data to only include Capomulin 
capomulin_data = clean_data[clean_data["Drug Regimen"] == "Capomulin"]

# only one mouse to be treated with capomulin
mouse_id = 's185'
mouse_data = capomulin_data[capomulin_data["Mouse ID"] == mouse_id]

# creation of the line plot for tumor volume vs. time point
plt.figure(figsize=(10, 6))
plt.plot(mouse_data["Timepoint"], mouse_data["Tumor Volume (mm3)"], marker='o')

# Setting plot labels and title
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Tumor Volume vs. Timepoint for Mouse {mouse_id} Treated with Capomulin")

# Showing the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
    #Capomulin was already declared above so should not duplicate

# average tumor volume for each mouse
average_tumor_vol = capomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().reset_index()

# mouse weight information
mouse_weight = capomulin_data[["Mouse ID", "Weight (g)"]].drop_duplicates() 

# Merging data on
merged_data = pd.merge(average_tumor_vol, mouse_weight, on="Mouse ID")

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_data["Weight (g)"], merged_data["Tumor Volume (mm3)"], color='b')

# Setting plot labels and title
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume for Capomulin Regimen")

# Showing the plot
plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# creation of variable to extract weight and average tumor volume data
weight = merged_data["Weight (g)"]
average_tumor_volume = merged_data["Tumor Volume (mm3)"]

# calculating the correlation coefficient
#reference https://www.geeksforgeeks.org/exploring-correlation-in-python/?ref=header_outind
correlation_coefficient, _ = st.pearsonr(weight, average_tumor_volume)
print(f"Correlation coefficient between mouse weight and average tumor volume: {correlation_coefficient:.2f}")


# Linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(weight, average_tumor_volume)

# Regression line equation
regression_line = slope * weight + intercept

# Creation of scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(weight, average_tumor_volume, color='b', label='Data')
plt.plot(weight, regression_line, color='r', label='Regression Line')

# Setting plot labels and title
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume for Capomulin Regimen")
plt.legend()

# Displaying the scatterplot
plt.show()
```
