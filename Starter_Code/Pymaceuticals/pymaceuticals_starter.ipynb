```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="inner")

# Display the data table for preview
merged_data
```

```python
# Checking the number of mice.
#create a variable for checking unique values in column
unique_mouse_id = pd.unique(merged_data['Mouse ID'])
# initiate the variable at 0 to start the count
# https://www.geeksforgeeks.org/python-get-unique-values-list/?ref=header_outind
mouse_unique = 0 
for item in unique_mouse_id:
    mouse_unique +=1
#printing the number of unique values within the group
print(f'Number of unique values present in the Mouse ID is: {mouse_unique}')
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
unique_timepoint = pd.unique(merged_data['Timepoint'])
timepoint_unique = 0
for item in unique_timepoint:
    timepoint_unique +=1
print(f'Number of unique values present in the Timepoint is: {timepoint_unique}')

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
    #https://www.geeksforgeeks.org/find-duplicate-rows-in-a-dataframe-based-on-all-or-selected-columns/?ref=header_outind
duplicate_mice = merged_data[merged_data.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
#use the loc function to select the first column

duplicate_mice = merged_data.loc[merged_data["Mouse ID"] == "g989", :]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#https://www.geeksforgeeks.org/python-pandas-dataframe-drop_duplicates/?ref=header_outind

clean_data = merged_data[merged_data["Mouse ID"] != "g989"]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame. (Reference to line 2 of the code)
#change the values associated with the previous block of code, adjusted to referenced dataframe and new module
#create a variable for checking unique values in column
mice_count = pd.unique(clean_data['Mouse ID'])
# initiate the variable at 0 to start the count
# https://www.geeksforgeeks.org/python-get-unique-values-list/?ref=header_outind
mouse_unique_clean = 0 
for item in mice_count:
    mouse_unique_clean +=1
#printing the number of unique values within the group
print(f'Number of unique values present in the Mouse ID in the mice count dataframe is: {mouse_unique_clean}')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.

    # https://www.geeksforgeeks.org/how-to-calculate-summary-statistics-in-pandas/?ref=header_outind <----calculation summary stats
        # https://www.geeksforgeeks.org/python-pandas-dataframe-aggregate/?ref=header_outind <----- aggregation method 
       # We would group the data by drug regimine so we can calculate the data for tumor volume, create new variable 

regimen_stats = clean_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean ='mean',
    Median ='median',
    Variance = 'var',
    Standard_Deviation = 'std',
    SEM = 'sem'
)    

# Assemble the resulting series into a single summary DataFrame.
    #to do this we would create a new variable that utilizes the pandas library to create a new dataframe that calls on the dataframe created above
#note to self "simpler way to do this would be to simply rename the dataframe above"

summary_stats = pd.DataFrame(regimen_stats)
summary_stats

#note to self "do not print both dataframes"
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
# For this section we will be utilizing the matplotlib library. information can be found in chapter 5.2-3
#this part is very important! reference the clean data that was generated in line 25
    #code should be structured as follows 1. new df name 2. equal 3.referenced dataset 4. referenced columns 5.function 6.by 7.function
    #Make sure your spelling is correct!

#grouping the data by drug regimen and counting
total_timepoints = clean_data[["Timepoint", "Drug Regimen"]].groupby(["Drug Regimen"]).count()

#sorting the data so we can plot it 
sorted_timepoints = total_timepoints.sort_values(["Timepoint"], ascending=False)
    #we have already created the variable so now we just need to apply a funciton to it
    #we use the function ascending = false to make sure the data is shown in decending order
#plotting the chart using dataframe.plot method
    # variable = referenced . function (plot) . type (bar) ()
        #made a mistake here and wrote .plt , should use .plot when utilizing matplotlib
timepoints_plot = sorted_timepoints.plot.bar()

#set the labels for the plot
timepoints_plot.set_ylabel("Timepoint")
timepoints_plot.set_xlabel("Drug Regimen")

timepoints_plot

# Educational reference for the above
    #https://www.geeksforgeeks.org/how-to-set-dataframe-column-value-as-x-axis-labels-in-python-pandas/?ref=header_outind
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
    #creating the same labels and graph using pyplot
plt.bar(sorted_timepoints.index, sorted_timepoints['Timepoint'], color='b', alpha=1, align='center')

#setting the axis labels
plt.ylabel('Timepoints')
plt.xlabel('Drug Regimen')
# The axis labels are horizontal by default so we should make them vertical for readability
plt.xticks(rotation='vertical')

# Displaying the graph
plt.show()
```

```python
# Generate a pie chart, using Pandas, showing the distribution of unique female versus male mice used in the study
unique_gender = clean_data.groupby(["Sex"]).count()
unique_gender_graph = unique_gender.plot.pie(y= "Mouse ID", title= "distribution of unique female versus male mice", autopct= "%1.2f%%")
unique_gender_graph.set_ylabel("Gender")
# Get the unique mice with their gender
unique_mice_gender = clean_data[['Mouse ID', 'Sex']].drop_duplicates()
print(unique_mice_gender)
# Make the pie chart
plt.show()
```

```python
# Generate a pie chart, using pyplot, showing the distribution of unique female versus male mice used in the study
unique_gender = clean_data.groupby(["Sex"])["Sex"].count()
plt.pie(unique_gender, autopct='%1.2f%%', labels= ('Female', 'Male'))


# Setting the title
plt.title("distribution of unique female versus male mice")
plt.legend(labels=('Female', 'Male'))
plt.ylabel('Gender')

# Displaying the graph
plt.show()
```
